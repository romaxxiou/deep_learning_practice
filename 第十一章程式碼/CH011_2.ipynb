```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# 
src = cv2.imread("Lenna.jpg")
kernel = np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]],dtype="float32")
kernel2=np.array([[-1,0,1],
                  [-2,0,2],
                  [-1,0,1]],dtype="float32")
# 
image = cv2.filter2D(src,-1,kernel)
image2= cv2.filter2D(src,-1,kernel2)
htich = np.hstack((src, image,image2))
plt.imshow(htich)
plt.show()
```
